```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def man_wom_chi(passenger):
    age=passenger['Age']
    sex=passenger['Sex']
    
    return 'child' if age < 16 else sex


df_data['Person'] = df_data.apply(man_wom_chi,axis=1)
```

```python
df_data[0:10]
```

```python
print(df_data['Person'].value_counts())
```

```python
sns.catplot(x='Pclass',data=df_data,hue='Person',kind='count')
```

```python
df_data['Age'].hist()
```

```python
fig = sns.FacetGrid(df_data,hue='Pclass',aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = df_data['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(df_data,hue='Sex',aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = df_data['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```
